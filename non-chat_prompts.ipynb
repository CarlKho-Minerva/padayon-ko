#### User-Facing Guidelines
# Debate SEXI Framework Guidelines

## State Your Point
1. **What is your main argument or position?**
   - **Identify the core message you want to convey:**
     - What is the one key idea you want to communicate?
     - How can you summarize this idea in one clear sentence?
   - **Formulate a precise statement:**
     - How can you rephrase this idea to make it more impactful?
     - Optional: Can you think of a slogan or catchphrase that captures this point?

## Explain Your Point
2. **Why is this point important?**
   - **Clarify the relevance:**
     - What are the consequences if this point is ignored?
     - How does this point address a current issue or need?
   - **Detail the implications:**
     - What specific problems or benefits are associated with this point?
     - Optional: Can you outline how this point impacts different stakeholders?

3. **What are the main reasons supporting your point?**
   - **List the supporting reasons:**
     - What are the key arguments or facts that back up your point?
     - Optional: How do these reasons interconnect to strengthen your argument?
   - **Provide evidence:**
     - What data or research supports these reasons?
     - Optional: Are there any expert opinions or studies you can cite?

## Illustrate Your Point
4. **Can you provide specific examples or evidence?**
   - **Choose relevant examples:**
     - What real-world instances or cases can you use to illustrate your point?
     - Optional: Are there any historical or current events that support your argument?
   - **Describe the examples:**
     - How do these examples demonstrate the validity of your point?
     - Optional: Can you include any visual aids or anecdotes to make your point clearer?
